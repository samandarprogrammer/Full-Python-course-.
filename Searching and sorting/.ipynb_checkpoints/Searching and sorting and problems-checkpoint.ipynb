```python
def pushZerosAtEnd(arr,n):
    i=0
    k=0
    while i<n and k<n:
        if arr[i]!=0 and arr[k]!=0:
            i=i+1
            k=k+1
        elif arr[i]==0:
            i=i+1
        else:
            arr[i],arr[k]=arr[k],arr[i]
            i=i+1
            k=k+1
    
    
n=int(input())
arr=[int(x) for x in input().strip().split(" ")]
pushZerosAtEnd(arr,n)
print(arr)
```

```python
def Rotate_array(arr,n,d):
    arr1=[]
    for i in range (0,d):
        arr1.append(arr[i])
    for j in range(0,n-d):
        arr[j],arr[j+d]=arr[j+d],arr[j]
    for k in range (0,d):
        arr[n-d+k],arr1[k]=arr1[k],arr[n-d+k]
        
n=int(input())
arr=[int(x) for x in input().strip().split(" ")]
d=int(input())
Rotate_array(arr,n,d)
print(arr)
```

```python
def Second_max(arr,n):
    l=-99999
    s=-99999
    for i in range(0,n):
        if arr[i]>l:
            s=l
            l=arr[i]
        if arr[i]>s and arr[i]!=l:
            s=arr[i]
    return s

n=int(input())
arr=[int(x) for x in input().strip().split(" ")]
print(Second_max(arr,n))
```

```python
def arrayRotateCheck(arr, n):
    count=0
    for i in range(n-1,1,-1):
        if arr[i]>arr[i-1]:
            pass
        else:
            count=count+1
            while i-1>0:
                count=count+1
                i=i-1
            return count
    return count

n=int(input())
arr=[int(x) for x in input().strip().split(" ")]
print(arrayRotateCheck(arr,n))
```

```python
def sort012(arr,n):
    nz=0
    nt=n-1
    for i in range(0,n):
        if arr[i]==1:
            pass
        elif arr[i]==0:
            arr[i],arr[nz]=arr[nz],arr[i]
            nz=nz+1
        else:    
            if arr[i]==2 and nt>i:
                arr[i],arr[nt]=arr[nt],arr[i]
                nt=nt-1
             
            if arr[i]==0:
                arr[i],arr[nz]=arr[nz],arr[i]
                nz=nz+1
                
n=int(input())
arr=[int(x) for x in input().strip().split(" ")]
sort012(arr,n)
print(arr)
```
