```python
def code16(n):
    i=1
    print(1)
    while i<n:
        j=0
        while j<=i:
            if j==0:
                print(i,end='')
            elif i==j:
                print(i,end='')
            else:
                print(0,end='')
            j=j+1
        print()    
        i =i+1
      
n=int(input())
code16(n)
```
